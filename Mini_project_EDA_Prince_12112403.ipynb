### Q5.To plot bar graph I used matplotlib library and plotted the bar graph of column  Marital_status and education vs count .

```python

selected_columns = ['Marital_Status', 'Education']

for column in selected_columns:
    category_counts = df[column].value_counts()
    category_counts.plot(kind='bar',color='Black',hatch='/',edgecolor='w')
    plt.xlabel(column)
    plt.ylabel('Count')

    plt.title('Bar Plot of {}'.format(column))

    plt.show()






```

### Q8.To select the numerical columns i used select_dtypes() fun and passed int or float dtype in it then i selected the columns in the list numerical_columns and i find the correlation using corr() fun on the numerical columns and i used  unstack() function is used to convert the correlation matrix into a Series object, making it easier to sort and extract the top correlated pairs. and after sorting in descending order  i selected top 3 values and ploted the scatter plot

```python

sel = df.select_dtypes('int'or'float').columns

numerical_columns = [col for col in sel]
corr_matrix=df[numerical_columns].corr()
positive_corr_matrix = corr_matrix.where(corr_matrix > 0)
#unstack() function is used to convert the correlation matrix into a Series object, making it easier to sort and extract the top correlated pairs.
top_corr_pairs = positive_corr_matrix.unstack().sort_values(ascending=False)[:3]
top_corr_columns = [x for x in top_corr_pairs.index]

for pair in top_corr_columns:
    plt.scatter(df[pair[0]], df[pair[1]])
    plt.xlabel(pair[0])
    plt.ylabel(pair[1])
    plt.title('Scatter Plot of {} vs {}'.format(pair[0], pair[1]))

    plt.show()
```

### Q9.geting the pairs of numerical columns having the top 3 negative correlation values and ploting the scatter plot of them

```python
sel = df.select_dtypes('int'or 'float').columns
print(sel)
numerical_columns = [col for col in sel]
corr_matrix=df[numerical_columns].corr()
negative_corr_matrix = corr_matrix.where(corr_matrix < 0)
#unstack() function is used to convert the correlation matrix into a Series object, making it easier to sort and extract the top correlated pairs.
top_corr_pairs = negative_corr_matrix.unstack().sort_values(ascending=True)[:3]
top_corr_columns = [x for x in top_corr_pairs.index]


for pair in top_corr_columns:
    plt.scatter(df[pair[0]], df[pair[1]])
    plt.xlabel(pair[0])
    plt.ylabel(pair[1])
    plt.title('Scatter Plot of {} vs {}'.format(pair[0], pair[1]))

    plt.show()
```
